## Data Fetching And Cleaning
* In this file we are going to do data fetching. Fetchimg images of cats and dogs from the internet using request and beatifulsoup.
* The images will be saved in different directories which will be `raw_cats` and `raw_dogs`

```python
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
cat_soup = bs(requests.get('https://www.google.com/search?q=cats+images&rlz=1C1CHZN_enZA940ZA940&sxsrf=ALeKk01oMopnNnj035lFJLuf2w3p9ESplQ:1619524073366&source=lnms&tbm=isch&sa=X&ved=2ahUKEwiuhOSorZ7wAhWWu6QKHQfsASoQ_AUoAXoECAEQAw&biw=1517&bih=741').content)
dog_soup = bs(requests.get('https://www.google.com/search?q=dogs&tbm=isch&ved=2ahUKEwiL5OyrrZ7wAhVM3hoKHaxfCUcQ2-cCegQIABAA&oq=dogs&gs_lcp=CgNpbWcQAzIECAAQQzIECAAQQzIECAAQQzIECAAQQzIHCAAQsQMQQzIECAAQQzIECAAQQzIECAAQQzIECAAQQzIECAAQQzoFCAAQsQNQi5YPWJacD2Ddng9oAHAAeACAAZAFiAHeCZIBAzUtMpgBAKABAaoBC2d3cy13aXotaW1nwAEB&sclient=img&ei=7_mHYMuwMMy8a6y_pbgE&bih=741&biw=1517&rlz=1C1CHZN_enZA940ZA940').content)
```

```python
dogs_folder_path = "raw_dogs"
cats_folder_path = "raw_cats"
```

#### Getting dogs images

```python
dogs = dog_soup.findAll('img')
dogs_src = [i["src"] for i in dogs]
dogs_src_filted = [i for i in dogs_src if re.match('^https?://', i)]
dogs_src_filted
```

#### Now we can download dogs images

```python
for i, dog in enumerate(dogs_src_filted):
    res = requests.get(dog)
    file = open(f'{dogs_folder_path}/dog{i+1}.png', 'wb')
    print(f"Downloading dog{i+1}.png")
    file.write(res.content)
    file.close()
```

#### Getting cats images

```python
cats = cat_soup.findAll('img')
cats_src = [i["src"] for i in cats]
cats_src_filted = [i for i in cats_src if re.match('^https?://', i)]
cats_src_filted
```

#### Downloading the cats image

```python
for i, cat in enumerate(cats_src_filted):
    res = requests.get(cat)
    file = open(f'{cats_folder_path}/cat{i+1}.png', 'wb')
    print(f"Downloading cat{i+1}.png")
    file.write(res.content)
    file.close()
```

### Let's convert the images of cats and dogs to gray images

> We want our gray images to have the same **shape** and it is important. (`200x200`)

```python
cats_path_gray = 'cats_gray'
dogs_path_gray = 'dogs_gray'
```

### 1. Dogs

```python
os.listdir(dogs_folder_path)[:2]
```

```python
dogs_images = [f'{dogs_folder_path}/{i}' for i in os.listdir(dogs_folder_path)]
for i, dog in enumerate(dogs_images):
    img_gray = cv2.imread(dog)
    
    img_gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
    img_gray = cv2.resize(img_gray,(200, 200))
    cv2.imwrite(f'{dogs_path_gray}/dog{i+1}.png', img_gray)
    print(f"SAVING dog{i+1}.png", img_gray.shape)
```

### 2. Cats

```python
os.listdir(cats_folder_path)[:2]
```

```python
cats_images = [f'{cats_folder_path}/{i}' for i in os.listdir(cats_folder_path)]
for i, cat in enumerate(cats_images):
    img_gray = cv2.imread(cat)
    
    img_gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
    img_gray = cv2.resize(img_gray,(200, 200))
    cv2.imwrite(f'{cats_path_gray}/cat{i+1}.png', img_gray)
    print(f"SAVING cat{i+1}.png", img_gray.shape)
```
